### Exploratory Data Analysis (EDA)

### Class Distribution

```python
class_counts = credit_df['Class'].value_counts()
class_counts
```

```python
plt.figure(figsize=(5,4))
sns.countplot(x='Class', data=credit_df)
plt.title("Class Distribution (0 = Legit, 1 = Fraud)")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

### Percentage of Fraud

```python
fraud_percentage = class_counts[1] / class_counts.sum() * 100
print(f"Fraud transactions: {fraud_percentage:.3f}%")
```

### Transaction Amount Distribution

```python
plt.figure(figsize=(6,4))
sns.histplot(credit_df['Amount'], bins=50, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

### Amount vs Class

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Class', y='Amount', data=credit_df)
plt.title("Transaction Amount by Class")
plt.show()
```

### Time Feature Analysis

```python
plt.figure(figsize=(6,4))
sns.histplot(credit_df['Time'], bins=50)
plt.title("Transaction Time Distribution")
plt.show()
```

### Correlation Analysis (Sampled)

```python
sample_df = credit_df.sample(50000, random_state=42)

plt.figure(figsize=(12,8))
sns.heatmap(
    sample_df.corr(),
    cmap="coolwarm",
    center=0,
    linewidths=0.1
)
plt.title("Feature Correlation Heatmap (Sampled)")
plt.show()
```

## EDA Summary – Credit Card Dataset

The dataset contains highly anonymized numerical features (V1–V28) derived from PCA.

No missing values or invalid data types were observed.

The target variable is extremely imbalanced, with fraudulent transactions representing less than 0.2% of the data.

Transaction amount and time exhibit distinct distributions between fraudulent and non-fraudulent classes.

Due to severe class imbalance, resampling techniques are required during model training.
